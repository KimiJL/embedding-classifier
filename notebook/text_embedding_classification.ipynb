```python
from embedding_classifier.embeddings import SentenceTransformerEmbeddings
from embedding_classifier.classifier import EmbeddingClassifier
from embedding_classifier.model import RandomForest
from datasets import load_dataset
```

```python
ds = load_dataset("imdb")
```

```python
train_valid = ds["train"].train_test_split(test_size=0.2)
```

```python
train_X = train_valid["train"]["text"]
train_Y = train_valid["train"]["label"]
valid_X = train_valid["test"]["text"]
valid_Y = train_valid["test"]["label"]
```

```python
DEVICE = "cuda"

embedder = SentenceTransformerEmbeddings("avsolatorio/GIST-small-Embedding-v0", device=DEVICE)
model =  RandomForest()
classifier = EmbeddingClassifier(embedding=embedder, model=model)
```

```python
classifier.fit(train_X, train_Y)
```

```python
act_pred = classifier.predict(valid_X)
```

```python
from sklearn.metrics import f1_score
f1_score(valid_Y, act_pred)
```
